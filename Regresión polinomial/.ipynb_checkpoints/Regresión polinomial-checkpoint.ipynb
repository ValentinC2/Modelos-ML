```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets, linear_model
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures
```

```python
data_url = "http://lib.stat.cmu.edu/datasets/boston"
raw_df = pd.read_csv(data_url, sep="\s+", skiprows=22, header=None)
data = np.hstack([raw_df.values[::2, :], raw_df.values[1::2, :2]])
target = raw_df.values[1::2, 2]
```

```python
x = data[:, np.newaxis,5]
x_train, x_test, y_train, y_test = train_test_split(x, target)
```

```python
reg_pol = PolynomialFeatures(degree=2)
x_train_pol = reg_pol.fit_transform(x_train)
x_test_pol = reg_pol.fit_transform(x_test)
```

```python
model = linear_model.LinearRegression()

model.fit(x_train_pol, y_train)
predict = model.predict(x_test_pol)
```

```python
print(x_train.shape)
print(y_train.shape)
print(predict.shape)
```

```python
plt.scatter(x_train, y_train)
plt.plot(x_test, predict, color="r")
plt.show
```

```python
model.score(x_test_pol, y_test)
```

```python
train=data[:, 6].reshape(-1, 1)

reg_pol = PolynomialFeatures(degree=2)

train_pol = reg_pol.fit_transform(train)

model.fit(train_pol , target)
predict = model.predict(train_pol)

plt.scatter(data[:, 6], target)
plt.plot(data[:, 6], predict, color="r")
plt.show()

model.score(train_pol, target)
```
