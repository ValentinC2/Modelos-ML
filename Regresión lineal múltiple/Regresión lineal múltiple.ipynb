```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets, linear_model
from sklearn.model_selection import train_test_split
```

```python
data_url = "http://lib.stat.cmu.edu/datasets/boston"
raw_df = pd.read_csv(data_url, sep="\s+", skiprows=22, header=None)
data = np.hstack([raw_df.values[::2, :], raw_df.values[1::2, :2]])
target = raw_df.values[1::2, 2]
```

```python
x_multiple = data[:, 5:8]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_multiple, target)
```

```python
model = linear_model.LinearRegression()

model.fit(x_train, y_train)
```

```python
predict = model.predict(x_test)
```

```python
plt.scatter(data[:, 5], target)
plt.scatter(data[:, 6], target)
plt.scatter(data[:, 7], target)
plt.plot(x_test, predict, color="r")
plt.show()
```

```python
model.score(x_test, y_test)
```

```python
train=data[:, 5].reshape(-1, 1)

model.fit(train , target)
predict = model.predict(train)

plt.scatter(data[:, 5], target)
plt.plot(data[:, 5], predict, color="r")
plt.show()

model.score(train, target)
```
